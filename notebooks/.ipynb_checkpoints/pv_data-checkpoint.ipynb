```python
import pandas as pd
import glob

files = glob.glob("/Users/vivekarya/Documents/GitHub/bess_trading_assessment/data/raw/demand_extracted/*.[cC][sS][vV]")

print("Files found:", len(files))

df_list = [pd.read_csv(f, header=None, low_memory=False) for f in files]

demand_raw = pd.concat(df_list, ignore_index=True)

print("Shape before removing duplicates:", demand_raw.shape)

demand_raw = demand_raw.drop_duplicates()

print("Shape after removing duplicates:", demand_raw.shape)

demand_raw.head(10)
```

```python
demand_raw.to_csv("demand_Actual.csv", index=False)
```
